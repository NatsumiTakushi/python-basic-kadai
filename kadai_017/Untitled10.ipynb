```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}さんは{self.age}歳なので、大人です。")
    else:
      print(f"{self.name}さんは{self.age}歳なので、大人ではありません。")

person1 = Human("侍 太郎", 25)
person2 = Human("侍 花子", 18)
person3 = Human("侍 三郎", 20)

human_list = [person1, person2, person3]

for human in human_list:
  human.check_adult()
```
