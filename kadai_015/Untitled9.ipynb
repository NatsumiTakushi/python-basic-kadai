```python
# Humanクラスの定義
class Human:
    # 初期化メソッド（コンストラクタ）
    def __init__(self, name, age):

        self.name = name
        self.age = age

    def printinfo(self):
        print(f"名前: {self.name}")
        print(f"年齢: {self.age}")

person = Human("侍 太郎", 20)

# メソッドの呼び出し
person.printinfo()
```
